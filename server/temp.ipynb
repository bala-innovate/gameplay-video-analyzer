```python
import yt_dlp
import shutil
from config import *
import os

video_url = 'https://youtu.be/U_91sJB-hwk?si=NbwetolieFEfvwz0'

node_abs = shutil.which('node')          # e.g. /usr/local/bin/node

        
ydl_opts = {
    'outtmpl': os.path.join(VIDEOS_DIR, '%(id)s.%(ext)s'),
    'format': 'bestvideo+bestaudio/best',
    'merge_output_format': 'mp4',
    'quiet': True,
    # 'verbose': True,           # extra debug
    'js_runtimes': {'node': {'executable': node_abs}},
}

try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(video_url, download=True)
        # ydl.download([self.video_url])
        video_path = ydl.prepare_filename(info)
    
        if not video_path.endswith('.mp4'):
            video_path = os.path.splitext(video_path)[0] + '.mp4'
        print(f'Video downloaded at location: {VIDEOS_DIR}')
except Exception as e:
    print(f"Error downloading video: {e}")
```
